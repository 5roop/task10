```python
import pandas as pd

df = pd.read_json("bcs_polsent_007.jsonl", orient="records", lines=True)
df.head(1)
```

```python
df.columns
```

```python
# Fix column ordering

cols_to_keep = [
    "sentence",
    "country",
    "origin",
    "annotator1",
    "annotator2",
    "reconciliation_hard",
    "label",
    "doc_id",
    "sentence_id",
    "date",
    "fullname",
    "party",
    "gender",
    "yob",
    "split"
]
 
```

```python
set(df.columns).difference(set(cols_to_keep))
```

 Next we keep only the columns we want to keep, which means we will drop columns `{'edu_y', 'gold', 'id_meta', 'ideology', 'no_seats', 'ruling', 'term', 'type'}`.

```python
# Rename columns to be as close as possible to the description in the paper.:
df = df[cols_to_keep].rename(columns={
    "origin": "round",
    "yob": "birth_year",
    "doc_id": "document_id",
    "reconciliation_hard": "reconciliation",
    "fullname": "name"
})
# Replace genders with explicit labels
df["gender"] = df.gender.replace(0, "M").replace(1, "F")

# Rename round to 'first' or 'second'

df["round"] = df["round"].replace("1-1300", "first").replace("1301-2600", "second")
```

```python
df.to_json("013_polsent.json", orient="records", lines=True)
```
